```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv2.imread('../motherforker-9000/images/sitterzaal/110cm.jpg',0)
img = cv2.flip(img, -1)

# show original image
plt.figure(figsize=(10,10))
plt.subplot(1, 1, 1)
plt.imshow(img)
plt.title('original')
plt.show()

# threshold image
img_threshold = cv2.threshold(img,70,255,cv2.THRESH_BINARY)[1]

# nb_components, output, stats, centroids = cv2.connectedComponentsWithStats(img)
# Without statistics
nb_components, output = cv2.connectedComponents(img_threshold, connectivity=8)
# Show connected components with random colours
colors = [np.random.randint(0,255,3) for _ in range(nb_components)]
colors[0] = [0, 0, 0] # for background
img_connected = np.zeros((output.shape[0], output.shape[1], 3), dtype=np.uint8)
for i in range(1, nb_components):
    img_connected[output == i] = colors[i]
plt.figure(figsize=(10,10))
plt.subplot(1, 1, 1)
plt.imshow(img_connected)
plt.title('connected components')
plt.show()
```
