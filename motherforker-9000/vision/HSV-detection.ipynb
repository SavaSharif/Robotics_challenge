```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

img = cv2.imread("blue.jpg")
# img = cv2.flip(img, -1)

# img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_blur = cv2.GaussianBlur(img_rgb,(5,5),0)
img_hsv = cv2.cvtColor(img_blur, cv2.COLOR_RGB2HSV)

min_hue = 110
max_hue = 130
min_sat = 50
max_sat = 255
min_val = 50
max_val = 255

# img_blur = cv2.GaussianBlur(img_rgb,(5,5),0) 
binary_img = cv2.inRange(img_hsv, (min_hue, min_sat, min_val), (max_hue, max_sat, max_val))

# edges = cv2.Canny(img_rgb,200,500)

plt.subplot(1, 1, 1)
plt.imshow(img_blur)
plt.show()

plt.subplot(1, 1, 1)
plt.imshow(binary_img, cmap='gray')
plt.show()

binary_img
```

```python
edgeList = np.argwhere(binary_img)

# y_mean = edgeList[:,0].mean() # vertical position is not really relevant
x_mean = edgeList[:,1].mean()
print("Mean x of edges:", x_mean)
```
