```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img = cv2.imread("../images/sitterzaal/110cm.jpg")
img = cv2.flip(img, -1)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

edges = cv2.Canny(img_rgb,200,500)
knipknip = edges[:,60:260]

plt.subplot(1, 1, 1)
plt.imshow(img_rgb)
plt.show()

plt.subplot(1, 1, 1)
plt.imshow(knipknip, cmap='gray')
plt.show()
```

## 1

```python
y = max(np.where(knipknip == 255)[0])
print("lowest edge pixel:", y)
# 92 = 110 cm
# 96 = 90 cm
# klop geen reet meer van!!!!!!!!!!!!
if y < 92:
    print("110 cm")
elif y > 96:
    print("90 cm")
else:
    d = 90 + 20 * (96 - y)/4
    print(int(d), "cm")
```

```python
edgeList = np.argwhere(knipknip)

# y_mean = edgeList[:,0].mean() # vertical position is not really relevant
x_mean = edgeList[:,1].mean()
print("Mean x of edges:", x_mean)

if x_mean < 140:
    print("left")
elif x_mean > 180:
    print("right")
else:
    print("gas op die lollie")
```
